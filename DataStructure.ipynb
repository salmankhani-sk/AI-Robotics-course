#  What is a Data Structure?

##  Definition
A **data structure** is a way to **store, organize, and manage data** in memory so that it can be **accessed and modified efficiently** during programming.

Data structures help in:
- Organizing data logically
- Improving code efficiency and performance
- Performing operations like search, insert, delete, sort, etc.

---

##  Types of Data Structures in Python

- Ordered, changeable, allows duplicates.
- Defined using square brackets `[]`.

#  What is Stack?

##  Definition
A **Stack** is a **linear data structure** that follows the **LIFO** (Last In, First Out) principle — the last element added is the **first one to be removed**.

---

##  Real-Life Example
- A stack of plates — the last plate placed on top is the first one you take off.

---

##  How Stack Works

| Operation   | Description                         |
|-------------|-------------------------------------|
| `push()`    | Adds (inserts) an element on top    |
| `pop()`     | Removes the top element             |
| `peek()`    | Returns the top element without removing it |
| `isEmpty()` | Checks if the stack is empty        |

---

##  Stack in Python (Using List)

```python
stack = []

# Push elements
stack.append(10)
stack.append(20)
stack.append(30)

# Pop element
stack.pop()  # Removes 30

# Peek top element
print(stack[-1])  # Output: 20

# Check if empty
print(len(stack) == 0)  # Output: False


````python
stack = ['z','x','y']
stack.append('a')
stack.append('b')
stack.append('c')
print(stack)
````

````python
last = stack.pop()
print(stack)
````

#  What is Queue?

##  Definition
A **Queue** is a **linear data structure** that follows the **FIFO** (First In, First Out) principle — the **first element added** is the **first one removed**.

---

##  Real-Life Example
- A line at the ticket counter — first person in line is served first.

---

##  How Queue Works

| Operation     | Description                          |
|---------------|--------------------------------------|
| `enqueue()`   | Add (insert) an element at the end   |
| `dequeue()`   | Remove an element from the front     |
| `peek()`      | View the front element               |
| `isEmpty()`   | Check if the queue is empty          |

---

##  Queue in Python (Using List – not recommended for large data)

```python
queue = []

# Enqueue elements
queue.append("A")
queue.append("B")
queue.append("C")

# Dequeue (remove from front)
queue.pop(0)  # Removes 'A'

# Peek front element
print(queue[0])  # Output: 'B'
```
